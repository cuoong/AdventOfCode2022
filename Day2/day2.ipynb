```python
file = open('../Day2/input.txt', 'r')
lines = file.readlines()
lines
```

```python
your_mapper = {
  "X": 1,
  "Y": 2,
  "Z": 3
}

score = 0

for item in lines:
  characters = item.strip().split(' ')
  your_choice = characters[1]
  opponent_choice = characters[0]
  
  if your_choice == "X" and opponent_choice == "A":
    score += (your_mapper[your_choice] + 3)
  elif your_choice == "X" and opponent_choice == "B":
    score += your_mapper[your_choice]
  elif your_choice == "X" and opponent_choice == "C":
    score += (your_mapper[your_choice] + 6)
  elif your_choice == "Y" and opponent_choice == "A":
    score += (your_mapper[your_choice] + 6)
  elif your_choice == "Y" and opponent_choice == "B":  
    score += (your_mapper[your_choice] + 3)
  elif your_choice == "Y" and opponent_choice == "C":
    score += your_mapper[your_choice]
  elif your_choice == "Z" and opponent_choice == "A":  
    score += your_mapper[your_choice]
  elif your_choice == "Z" and opponent_choice == "B":  
    score += (your_mapper[your_choice] + 6)
  elif your_choice == "Z" and opponent_choice == "C":
    score += (your_mapper[your_choice] + 3)
print(score)
```

```python
# X means you need to lose
# Y means you need to draw
# Z means you need to win
expection_mapper = {
  "X": 1,
  "Y": 2,
  "Z": 3
}
score = 0

for item in lines:
  characters = item.strip().split(' ')
  your_expectation = characters[1]
  opponent_choose = characters[0]
  
  if your_expectation == "X": # lose
    if opponent_choose == "A":
      score += expection_mapper["Z"]
    elif opponent_choose == "B":
      score += expection_mapper["X"]
    else:
      score += expection_mapper["Y"]
  elif your_expectation == "Y": # draw
    if opponent_choose == "A":
      score += (expection_mapper["X"] + 3)
    elif opponent_choose == "B":
      score += (expection_mapper["Y"] + 3)
    else:
      score += (expection_mapper["Z"] + 3)
  elif your_expectation == "Z": # win
    if opponent_choose == "A": # choose Rock
      score += (expection_mapper["Y"] + 6)
    elif opponent_choose == "B": #choose Paper
      score += (expection_mapper["Z"] + 6)
    else: # choose Scissors
      score += (expection_mapper["X"] + 6)
print("This is score {}".format(score))
```
