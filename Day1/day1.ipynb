```python
file = open('../Day1/input.txt', 'r')
lines = file.readlines()
lines
```

```python
data = []
sum = 0
for line in lines:
  line = line.strip()
  if line == '':
    data.append(sum)
    sum = 0
  else:
    sum += int(line)
print(max(data))
```

```python
total = 0
res = sorted(data)[-3:]
for number in res:
  total += number
print(total)
```
